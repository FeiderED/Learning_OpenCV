### Guarden una nueva imagen que contenga solo los canales rojo y verde:

```python
image = cv2.imread('paisaje.jpg')

red_green = image.copy()
red_green[:, :,0] = 0

cv2.imwrite('Rojo_Verde.jpg', red_green)
cv2.waitKey(0)
```

### Guarden una nueva imagen que muestre la imagen en escala de grises

```python
image = cv2.imread('paisaje.jpg')

gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

cv2.imwrite('grises.jpg', gray)
cv2.waitKey(0)
```
