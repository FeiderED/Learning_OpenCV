```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

## Cambien el tipo de color a RGB

```python
#Leer la imagen
img1 = cv2.imread('img/imagen1.jpg')
img2 = cv2.imread('img/imagen2.jpg')

#Resize de la imagen
img1 = cv2.resize(img1, (700, 500))
img2 = cv2.resize(img2, (700, 500))

#Cambiar el color de la imagen a RGB
imagen1_RGB = cv2.cvtColor(img1, cv2.COLOR_BGR2RGB)
imagen2_RGB = cv2.cvtColor(img2, cv2.COLOR_BGR2RGB)

#Mostrar la imagen
plt.imshow(imagen1_RGB)
plt.show()
plt.imshow(imagen2_RGB)
plt.show()
```

## Hagan girar una imagen

```python
#Rotacin de la imagen
rotacion180 = cv2.rotate(imagen1_RGB, cv2.ROTATE_180)
rotacion90 = cv2.rotate(imagen1_RGB, cv2.ROTATE_90_CLOCKWISE)


#Calcular las dimensiones de la imagen
(altura, ancho)=imagen1_RGB.shape[:2] #Forma de obtener las dimensiones de la imagen
(cX, cY) = (ancho//2, altura//2) #Centro de la imagen


#Matriz de rotacin 30 grados
M = cv2.getRotationMatrix2D((cX, cY), 30, 1.0) 
rotacion30 = cv2.warpAffine(imagen1_RGB, M, (ancho, altura))


#Mostrar la imagen rotada
plt.title('Rotacin 180')
plt.imshow(rotacion180)
plt.show()
plt.title('Rotacin 90')
plt.imshow(rotacion90)
plt.show()
plt.title('Rotacin 30 con matriz de rotacin')
plt.imshow(rotacion30)
plt.show()
```
